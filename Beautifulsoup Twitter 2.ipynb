```python
#Finally 
#to get all tweets in one page
import matplotlib.pyplot as plt
import urllib
import urllib.request
from bs4 import BeautifulSoup
theurl="http://twitter.com/realDonaldTrump"
thepage=urllib.request.urlopen(theurl)
soup=BeautifulSoup(thepage,"html.parser")
print(soup.title.text)

list=['']
#print(soup.find('div',{"class":"ProfileHeaderCard"}).find('p').text)
i=1
for tweets in soup.findAll('div',{"class":"content"}):
    
    # print(i)
    list.append(tweets.find('p').text)
    i=i+1
 
j=0
for j in range(0,i,1): 
  print(j,list[j])
  print('\n')
  
noftweets=i
#Analysis
print('----------------------------------Analysis------------------------------------------------------')

pos=0
neg=0
neut=0
final=[]
final.append(Analyze_Sentence(pos,neg,neut,noftweets))
print('Returned Values:',final)

# Data to plot
labels = 'Positive tweets', 'Negative Tweets', 'Neutral Tweets'
sizes = final[0]
colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.1, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
print(plt.show())


Total=final[0][0]+final[0][1]+final[0][2]
print('Total no of Tweets:',Total)
print('No of positive Tweets:',final[0][0],'  The Percentage of the positive Tweets : ',((final[0][0])*100)/Total)
print('No of negative Tweets:',final[0][1],'  The Percentage of the Negative Tweets : ',((final[0][1])*100)/Total)
print('No of neutral Tweets:',final[0][2],'  The Percentage of the Negative Tweets : ',((final[0][2])*100)/Total)




 

```

```python
'''
#Sentiment Analysis
The sentiment property returns a named tuple of the form Sentiment (polarity, subjectivity). 
The polarity score is a float within the range [-1.0, 1.0]. The subjectivity is a float
within the range [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective.

'''

from textblob import TextBlob
def Analyze_Sentence(pos,neg,neut,noftweets):
 
   for i in range(0,noftweets,1):
     analysis=TextBlob(list[i]).sentiment[0]
     if analysis > 0 :
         print('Positive')
         pos+=1
     elif analysis < 0:
         print('Negative')
         neg+=1
     else:
         print('Neutral')
         neut+=1
   a=[pos,neg,neut] 
   return a
```
